****Sales Data Cleaning and KPIs Calculations****

***Import and Clean the Data***

```python
#Loading the Data
import pandas as pd 
df = pd.read_csv('Week-2-Sales-Data.csv')
df.head()
```

```python
#Checking for missing values, and incorrect data types.
df.info()
```

```python
#Checking for Duplicate values
df.duplicated().sum()
```

```python
#Removing Duplicate values
df.drop_duplicates(inplace = True)
```

```python
#Converting 'Order_Date' to datetime format
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
#Checking the DataFrame info again
df.info()
```

*****Key Performance Indicators*****

```python
#a) Total Revenue
total_revenue = df['Revenue'].sum()
print("a) Total Revenue: R", total_revenue)
```

```python
#b) Average Units Sold per Order
avg_units_sold = df['Units_Sold'].mean() 
print("b) Average Units Sold per Order:", avg_units_sold)
```

```python
#c) Total Revenue per Region
revenue_per_region = df.groupby('Region')['Revenue'].sum() 
print("c) Total Revenue per Region:") 
print(revenue_per_region)
```

```python
#d) Highest Revenue-Generating Sales Rep
top_sales_rep = df.groupby('Sales_Rep')['Revenue'].sum().idxmax() 
top_sales_rep_revenue = df.groupby('Sales_Rep')['Revenue'].sum().max() 
print("d) Highest Revenue-Generating Sales Rep:", top_sales_rep, "with revenue:", top_sales_rep_revenue)
```

```python
#e) Top 3 Products by Units Sold
top_products = df.groupby('Product')['Units_Sold'].sum().sort_values(ascending=False).head(3)
print("e) Top 3 Products by Units Sold:")
print(top_products)
```
